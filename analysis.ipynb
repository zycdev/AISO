```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font',family='Times New Roman')
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
x  = [10,       20,    30,    50,   100,   300,   500,   700,  1000,  1500, 2000]
y1 = [82.16, 83.34, 83.92, 84.38, 84.88, 85.42, 85.51, 85.60, 85.69, 85.78, 85.82]
y2 = [73.44, 74.47, 75.04, 75.35, 75.71, 76.12, 76.26, 76.32, 76.45, 76.47, 76.49]
y3 = [ 3.24,  4.11,  4.84,  6.11,  8.74, 16.17, 22.57, 28.54, 36.64, 46.43, 54.50]
y4 = [721,     574,   508,   432,   339,   249,   228,   211,   190,   126,   116]
```

```python
sns.set_style('white')  # white whitegrid ticks
```

```python
# -  --   -.  :    ,
# .  ,   o   v    <    *    +    1
plt.plot(x, y1, label="P EM", linestyle="-", marker='v')
plt.plot(x, y2, label="Ans F1", linestyle="-", marker='s')
plt.plot(x, y3, label="Avg. # read", linestyle = "-", marker='o')
plt.xlabel("T")
plt.ylabel("Metric")
plt.xscale('log')
plt.grid(True)
# plt.title('sin & cos')
plt.legend()
plt.savefig('figs/step_metric.pdf', dpi=100)
plt.show()
```

```python
fig, ax1 = plt.subplots()
plt.xticks(rotation=60)

color = 'C2'
ax1.set_xlabel('The maximum number of steps $T$')
ax1.set_ylabel('Performance', color=color)
ax1.plot(x, y1, label="P EM", color=color, linestyle="--", marker='v')
ax1.plot(x, y2, label="Ans F1", color=color, linestyle=":", marker='s')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'C1'
ax2.set_ylabel('Cost', color=color)  # we already handled the x-label with ax1
ax2.plot(x, y3, label="Avg. # read", color=color, linestyle="-", marker='o')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='lower right')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.xscale('log')
plt.xticks(x, x)
# plt.grid(True)
plt.savefig('figs/step_metric.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
